```python
import tensorflow as tf
import tensorflow_addons as tfa
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
img_raw = tf.io.read_file("tucano.jpg")
img = tf.io.decode_image(img_raw)
img = tf.image.convert_image_dtype(img, tf.float32)
img = tf.image.resize(img, [500,500])

plt.imshow(img)
```

```python
sobel = tf.image.sobel_edges(img)
_ = plt.imshow(sobel)
```

## Translate

```python
translate = tfa.image.translate(img, [100, -50])
_ = plt.imshow(translate)
```

## Rotate

```python
rotate = tfa.image.rotate(img, tf.constant(np.pi/3))
_ = plt.imshow(rotate)
```

## Scale

```python
scale = tf.image.random_crop(img, [250, 250, 3])
scale = tf.image.resize(scale, [500,500])
_ = plt.imshow(scale)
```

## Flip

### Vertical Flip

```python
flip1 = tf.image.flip_up_down(img)
_ = plt.imshow(flip1)
```

### Horizontal Flip

```python
flip2 = tf.image.flip_left_right(img)

_ = plt.imshow(flip2)
```

### Random Vertical Flip

```python
flip3 = tf.image.random_flip_up_down(img)
_ = plt.imshow(flip3)
```

## Crop

```python
crop = tf.image.random_crop(img, [250, 250, 3])
_ = plt.imshow(crop)
```

## Noise

```python
noiseG = tf.random.normal(shape=tf.shape(img), mean=0.0, stddev=0.04, dtype=tf.float32)
noise = tf.add(img, noiseG)
_ = plt.imshow(noise)
```

## Brightness

```python
brightness = tf.image.adjust_brightness(img, 0.6)
_ = plt.imshow(brightness)
```

## Shear

```python
level = 0.3
```

### Shear X

```python
shearX = tfa.image.transform(img, [1., level, 0., 0., 1., 0., 0., 0.])
_ = plt.imshow(shearX)
```

### Shear Y

```python
shearY = tfa.image.transform(img, [1., 0., 0., level, 1., 0., 0., 0.])
_ = plt.imshow(shearY)
```

## Interpolation

```python
rotate = tfa.image.rotate(img, tf.constant(np.pi/3), interpolation='NEAREST')
_ = plt.imshow(rotate)
```
